```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_encoder = df.copy()
```

```python
for col in df_encoder.select_dtypes(include = "object").columns:
    df_encoder[col]=LabelEncoder().fit_transform(df_encoder[col])
```

```python
x = df_encoder.drop("placed",axis = 1)
y = df_encoder["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 9)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
```

```python
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100,pr.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(4,3))
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = pr) 
```
